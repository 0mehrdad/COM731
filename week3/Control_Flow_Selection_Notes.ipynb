# Relational Operators

- <font size ="3">Operators that specifies a **comparison** between two operands</font>
- <font size ="3">The operands may be two values, two variables, or one of each</font>
- <font size = "3">Boolean expressions: expressions formed from relational operators and operands that evaluate to **True** or **False**</font><br>
- <font size = "3">Relational operators include:</font>
    - <font size = "3">== (equal to)</font>
    - <font size = "3">!= (not equal to)</font>
    - <font size = "3">$<$ (less than)</font>
    - <font size = "3">$>$  (greater than)</font>
    - <font size = "3">$<=$ (less than or equal to)</font>
    - <font size = "3">$>=$  (greater than or equal to)</font>

```python
#Relational Operators Example 2
number_1 = 25
number_2 = 78

number_1 > number_2
```

# Selection Control/ Decision Control

# Introduction

- <font size= "3">A control structure that **alters the sequential flow** by branching to specific statements based on a **condition or decision**</font>
- <font size= "3">We use relational operators to set a **condition or decision**</font>
- <font size= "3">Decision control structures help programs seem intelligent because they can make responses that correspond to user input</font>
- <font size= "3">Allow programs to handle data based on differing values</font>

## **1.if-else statement**

- <font size="3">A decision coded as <b>two possible paths</b></font>
- <font size="3">Specify what statements should be executed when the condition of an if statement is evaluated as <span style="color: green;"><b>True</b></span></font>
- <font size="3">The <span style="color: green;"><b>else</b></span> keyword designates the alternative execution path</font>
- <font size="3">We can only have at most <b>one else with each if statement</b></font>
- <font size="3"><b>Syntax:</b></font><br>
&emsp;&emsp;&emsp;&emsp;<font size="3"> <span style="color: green;"><b>if</b></span> condition<b><span style="color: green;">:</span></b><br>
&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;<font size="3">sequence of statements-1<br>
&emsp;&emsp;&emsp;&nbsp;<font size="3"> <span style="color: green;"><b>else</b></span><b><span style="color: green;">:</span></b><br>
&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;<font size="3">sequence of statements-2<br>

**if-else statement example**

```python
#Check the eligibility to vote in England and Northern Ireland

# A person in England and Northern Ireland can vote when they are 18 years old

eligible_age = 18

# Retieve age from a user
elector_age = int(input("Please enter your age: "))

# Determine whether an individual is eligible to vote based on their age
if elector_age >= 18:
    print("You are eligible to vote!")
else:
    print("Sorry, you cannot vote.")
```

# 2. if-elif-else statement

- <font size="3">A structure used to handle algorithms that require <b>multiple conditions<b></font>
- <font size="3">The computer evaluates statements **only until one of the conditions is true**, then skips any remaining <b>elif</b> statements</font>
- <font size="3">The <b>elif</b> is optional and you can have as many as you like</font>
- <font size="3"><b>Syntax:</b></font><br>
&emsp;&emsp;&emsp;&emsp;<font size="3"> <span style="color: green;"><b>if<b></span> condition-1<b><span style="color: green;">:</span></b><br>
&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;<font size="3">sequence of statements-1<br>
&emsp;&emsp;&emsp;&nbsp;<font size="3"> <span style="color: green;"><b>elif</b></span> subsequent condition<b><span style="color: green;">:</span></b><br>
&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;<font size="3">sequence of statements-n<br>
&emsp;&emsp;&emsp;&nbsp;<font size="3"> <span style="color: green;"><b>else</b></span><b><span style="color: green;">:</span></b><br>
&emsp;&emsp;&emsp;&emsp;&emsp;&emsp;<font size="3">default sequence of statements<br>

### Example of if-elif-else statement

```python
'''Program: Check data scientist salary in the UK by experience
   Description: Determine average data scientist salary in the UK based on experience level 
   where higher experience results in a higher salary package.
   
   The table shows experience levels and corresponding average salaries.
   | Level          | Average Salary per annum
     Entry-Level      £39,521
     Mid-Level        £59,454
     Experienced      £80,878
   
   reference: https://www.simplilearn.com/data-scientist-salary-in-uk-article
'''

# Retrieve experience level
level = input("Enter your experience level: ")

# check the average salary by experience level 
if level == 'Entry-Level':
    avg_salary = 39521
elif level == 'Mid-Level':
    avg_salary = 59454
elif level == 'Experienced':
    avg_salary = 80878
else: 
    print("Unrecognised experience level")
    avg_salary = 0

if avg_salary == 0:
    print("Average salary cann't identified")
else:
    print(f"The {level} data scientist earns £{avg_salary:,} per annum. ")
```

## 3. **Multiple conditions**

- <font size="3">We can implement decisions with multiple condition using logical: <b>and</b>, <b>or</b>, and <b>not</b> operators</font>
    - <font size="3">Combine the outcomes of two or more Boolean expressions</font>
    - <font size="3">The logical operators work by evaluating expressions from the left to the right</font>
    - <font size="3">When writing a Boolean expression with multiple logical operators, use parentheses to ensure the logic is carried out in the correct sequence</font>

## <span style="color: green;">and</span> operator

- <font size="3">The <span style="color: green;">and</span> operator requires that <b>both</b> conditions be <b>true</b> for the expression to evaluate to <b>True</b></font>

**and operator example**

```python
'''Program to identify the IQ Classification.
   IQ classification is the practice of categorising human intelligence, as measured by intelligence quotient (IQ) tests.
   The Stanford–Binet scales was developed by Gale H. Roid and published in 2003 by Riverside Publishing.
   | IQ Range     | IQ Classification
     140+           Highly advanced
     130-140        Very advanced
     120-129        Superior
     110-119        High average
     90-109         Average
     80-89          Low average
     70-79          Borderline impaired
     55-69          Midly impaired
     40-54          Moderately impaired
'''

# Retieve the IQ range from a user
iq_range = int(input("Enter an IQ range: "))

# check the IQ classification with the IQ range
if (iq_range>140):
    print("Highly advanced")
elif (iq_range>=130) and (iq_range<=140):
    print("Very advanced")
elif (iq_range>=120) and (iq_range<=129):
    print("Superior")
elif (iq_range>=110) and (iq_range<=119):
    print("High average")
elif (iq_range>=90) and (iq_range<=109):
    print("Average")
elif (iq_range>=80) and (iq_range<=89):
    print("Low average")
elif (iq_range>=70) and (iq_range<=79):
    print("Borderline impaired")
elif (iq_range>=55) and (iq_range<=69):
    print("Midly impaired")
elif (iq_range>=40) and (iq_range<=54):
    print("Moderately impaired")
else:
    print("Unidentified IQ Classification")
```

## <span style="color: green;">or</span> operator

- <font size="3">The or operator requires that <b>only one</b> of the conditions is <b>true</b></font>

**or operator example**

```python
'''Program to check a shape of object. '''

# Retrieve shape from user
print("What shape do I have?")
shape = input()
    
# Identify the shape
if (shape == "square") or (shape == "rectangle"):
    print("This shape is a parallelogram.")
else:
    print("This shape is not a parallelogram.")
```
